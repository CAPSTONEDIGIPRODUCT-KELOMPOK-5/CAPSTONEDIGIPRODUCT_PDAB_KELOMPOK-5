###**Outliers Values**

````python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    percent_outliers = (len(outliers) / len(df)) * 100
    results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)

display(results_df)
````

Kesimpulan yang dapat kita ambil dari persentase outliers untuk setiap kolom adalah sebagai berikut:

1. Kolom 'pm10': Persentase outliers pada kolom 'pm10' adalah sekitar 0.88%. Ini menunjukkan bahwa sebagian kecil dari nilai dalam kolom 'pm10' di luar jangkauan nilai yang dianggap normal atau mungkin tidak sesuai dengan pola umum data.

2. Kolom 'so2': Persentase outliers pada kolom 'so2' adalah sekitar 2.77%. Ini menunjukkan bahwa sekitar 2.77% dari data dalam kolom 'so2' dianggap sebagai outliers. Hal ini mungkin menunjukkan adanya variasi yang signifikan dalam data terkait dengan tingkat SO2 dalam udara.

###**Pre-Processing**

####**Validasi Silang**

````python
def kmeans(X, k):
    # Initialize centroids using k-means++
    centroids = [X[np.random.choice(len(X))]]
    while len(centroids) < k:
        distances = np.array([min(np.linalg.norm(x - c) ** 2 for c in centroids) for x in X])
        probabilities = distances / distances.sum()
        next_centroid = X[np.random.choice(len(X), p=probabilities)]
        centroids.append(next_centroid)
    centroids = np.array(centroids)

    distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))
    labels = np.argmin(distances, axis=0)
    new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])

    while not np.allclose(centroids, new_centroids):  # Iterate until convergence
        centroids = new_centroids
        distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))
        labels = np.argmin(distances, axis=0)
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])

    inertia = np.sum(np.min(distances, axis=0))
    return inertia, labels

# Generate example data for testing
np.random.seed(0)
X = np.random.rand(100, 2)

inertia_values = []
k_range = range(2, 13)

for k in k_range:
    inertia, _ = kmeans(X, k)
    inertia_values.append(inertia)

# Plot elbow method
plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.xticks(k_range)
plt.show()
````

Output diatas merupakan implementasi metode siku (elbow method) untuk menentukan jumlah optimal dari klaster dalam algoritma KMeans. Plot diatas akan menunjukkan bagaimana nilai inertia berubah seiring dengan jumlah klaster. Tujuan dari metode siku adalah untuk menemukan titik "siku" di plot, yaitu titik dimana penurunan nilai inertia tidak lagi signifikan saat jumlah klaster ditambahkan. Titik siku ini menunjukkan jumlah klaster optimal. Berdasarkan hasil plot diatas, dapat kita lihat jika titik "siku" berada di angka 6. Dapat kita simpulkan jika jumlah klaster optimal dalam algoritma KMeans adalah 6.

###**Build Model**

####**K-Means**

````python
import pandas as pd
import numpy as np

# Menentukan features
features = df2[['pm10', 'so2', 'co', 'o3', 'no2']]

# Menghitung mean dan std dari features
mean_features = features.mean()
std_features = features.std()

# Ganti std yang 0 dengan 1 untuk mencegah pembagian dengan nol
std_features[std_features == 0] = 1

# Normalisasi features
df_norm = (features - mean_features) / std_features

# Cek apakah ada nilai infinite setelah normalisasi dan gantikan dengan nol
df_norm.replace([np.inf, -np.inf], np.nan, inplace=True)
df_norm.dropna(inplace=True)  # Menghapus baris yang mengandung NaN

# Menentukan jumlah clusters
k = 6
# Inisialisasi centroids
centroids = df_norm.sample(k, random_state=42)

max_iterations = 100
for i in range(max_iterations):
    distances = pd.DataFrame(index=df_norm.index)
    for j in range(k):
        # Menghitung jarak Euclidean dari setiap titik ke centroid
        dist = np.sum((df_norm - centroids.iloc[j])**2, axis=1)
        distances[f'distance_to_{j}'] = dist.astype(float)  # Pastikan semua data adalah float

    # Menentukan label cluster terdekat
    cluster_labels = distances.idxmin(axis=1)
    cluster_labels = cluster_labels.map(lambda x: int(x.lstrip("distance_to_")))

    # Menghitung centroids baru
    new_centroids = pd.DataFrame(index=range(k), columns=df_norm.columns)
    for j in range(k):
        new_centroids.iloc[j] = df_norm[cluster_labels == j].mean()

    # Memeriksa konvergensi
    if centroids.equals(new_centroids):
        break
    centroids = new_centroids.copy()

# Menambahkan label cluster ke data asli
df2 = df2.loc[df_norm.index]  # Sesuaikan df2 untuk menyinkronkan dengan df_norm
df2["cluster"] = cluster_labels

# Ganti nama kolom 'cluster' menjadi 'kmeans_label'
df2.rename(columns={'cluster': 'kmeans_label'}, inplace=True)

print("\nCluster sizes:")
print(df2["kmeans_label"].value_counts())
````

Source code diatas merupakan implementasi Build Model menggunakan algoritma KMeans. Pada kode ini, kita akan membuat objek model KMeans dengan menggunakan jumlah klaster yang telah ditentukan dari perhitungan validasi silang (elbow method), yaitu 6 klaster 'n_clusters=6'.

####**K-Nearest Neighbors**

````python
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.base import BaseEstimator, ClassifierMixin

# Custom KNN Classifier
class KNNClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, k=5):
        self.k = k

    def fit(self, X, y):
        self.X_train_ = np.array(X)
        self.y_train_ = np.array(y)
        return self

    def predict(self, X):
        X = np.array(X)  # Ensure compatibility with numpy operations
        y_pred = []
        for x in X:
            # Calculate distances from x to all points in X_train_
            distances = np.sqrt(np.sum((self.X_train_ - x) ** 2, axis=1))
            # Get indices of the k nearest neighbors
            nearest_neighbors = np.argsort(distances)[:self.k]
            # Determine the majority class among the nearest neighbors
            majority_class = np.argmax(np.bincount(self.y_train_[nearest_neighbors]))
            y_pred.append(majority_class)
        return np.array(y_pred)

    def predict_proba(self, X):
        X = np.array(X)  # Ensure compatibility with numpy operations
        proba = []
        for x in X:
            # Calculate distances from x to all points in X_train_
            distances = np.sqrt(np.sum((self.X_train_ - x) ** 2, axis=1))
            # Get indices of the k nearest neighbors
            nearest_neighbors = np.argsort(distances)[:self.k]
            # Count occurrences of each class among the nearest neighbors
            class_counts = np.bincount(self.y_train_[nearest_neighbors], minlength=np.max(self.y_train_)+1)
            # Calculate probabilities for each class
            class_proba = class_counts / self.k
            proba.append(class_proba)
        return np.array(proba)

# Define the preprocessing for numerical features
num_pipeline = Pipeline(
    steps=[
        ('num_imputer', SimpleImputer()),  # Imputes missing values using the mean
        ('scaler', StandardScaler()),     # Scales features using standardization
    ]
)

# Define the preprocessing for categorical features
cat_pipeline = Pipeline(
    steps=[
        ('cat_imputer', SimpleImputer(strategy='most_frequent')),  # Imputes missing values using the most frequent value
        ('onehot', OneHotEncoder()),  # Applies one-hot encoding
    ]
)

# Define the column transformer for preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num_pipeline', num_pipeline, num_features),
        ('cat_pipeline', cat_pipeline, cat_features),
    ]
)

# Define the full pipeline including preprocessing and the classifier
knn_pipeline = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('classifier', KNNClassifier(k=4)),  # You can adjust k as needed
    ]
)

# Fit the model using the full pipeline
knn_pipeline.fit(X_train, y_train)

# Example prediction:
hasil_knn = knn_pipeline.predict(df5)
````

####**Gaussian Naive Bayes**

````python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.metrics import make_scorer

class MyGaussianNB(BaseEstimator, ClassifierMixin):
    def __init__(self, var_smoothing=1e-9):
        self.var_smoothing = var_smoothing

    def fit(self, X, y):
        # Calculate class priors
        self.class_prior_ = np.bincount(y) / len(y)
        self.class_count_ = np.bincount(y)

        # Calculate mean and standard deviation for each feature and class
        self.theta_ = np.zeros((len(np.unique(y)), X.shape[1]))
        self.sigma_ = np.zeros((len(np.unique(y)), X.shape[1]))
        for i, label in enumerate(np.unique(y)):
            X_class = X[y == label]
            self.theta_[i] = np.mean(X_class, axis=0)
            self.sigma_[i] = np.std(X_class, axis=0)

    def _calculate_likelihood(self, X):
        # Calculate likelihood for each class and sample
        likelihood = np.zeros((X.shape[0], len(self.class_prior_)))
        for i in range(len(self.class_prior_)):
            # Calculate likelihood using Gaussian distribution
            likelihood[:, i] = np.prod(
                np.exp(-0.5 * ((X - self.theta_[i]) / self.sigma_[i]) ** 2) /
                (np.sqrt(2 * np.pi) * self.sigma_[i]), axis=1)
        return likelihood

    def log_likelihood(y_true, y_pred_proba):
        log_likelihood = np.sum(np.log(y_pred_proba[np.arange(len(y_true)), y_true]))
        return log_likelihood


    def predict_proba(self, X):
        # Calculate posterior probabilities
        likelihood = self._calculate_likelihood(X)
        posterior = likelihood * self.class_prior_
        # Normalize to get probabilities
        return posterior / np.sum(posterior, axis=1)[:, np.newaxis]

    def predict(self, X):
        return np.argmax(self.predict_proba(X), axis=1)

# Define the preprocessing function
def preprocess_data(X):
    num_imputer = SimpleImputer()
    X_num = num_imputer.fit_transform(X)
    return X_num

# Combine preprocessing with Gaussian Naive Bayes classifier
gnb_pipeline = Pipeline(
    steps=[
        ('preprocessor', FunctionTransformer(preprocess_data)),
        ('classifier', MyGaussianNB()),
    ]
)

# Train the model using the pipeline
gnb_pipeline.fit(X_train, y_train)

# Perform prediction
hasil_gnb = gnb_pipeline.predict(df5)
````

###**ROC Curve**

````python
from sklearn.metrics import roc_auc_score, roc_curve
import numpy as np

models = [KNNClassifier(), MyGaussianNB()]
model_names = ['K-Nearest Neighbors', 'Gaussian Naive Bayes']

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for model, name, ax in zip(models, model_names, axes):
    model.fit(X_train, y_train)
    y_pred_proba = model.predict_proba(X_test)

    if len(np.unique(y_train)) > 2:  # Multiclass classification
        unique_classes = np.unique(y_train)
        for i in range(len(unique_classes)):
            fpr, tpr, _ = roc_curve(y_test == unique_classes[i], y_pred_proba[:, i])
            roc_auc = roc_auc_score(y_test == unique_classes[i], y_pred_proba[:, i])
            ax.plot(fpr, tpr, label=f'Class {unique_classes[i]} (ROC-AUC = {roc_auc:.2f})')

    else:  # Binary classification
        fpr, tpr, _ = roc_curve(y_test, y_pred_proba[:, 1])
        roc_auc = roc_auc_score(y_test, y_pred_proba[:, 1])
        ax.plot(fpr, tpr, label=f'{name} (ROC-AUC = {roc_auc:.2f})')

    ax.plot([0, 1], [0, 1], color='navy', linestyle='--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'ROC Curve - {name}')
    ax.legend(loc='lower right')

plt.tight_layout()
plt.show()
````

Pada ROC Curve diatas, dapat kita simpulkan lagi bahwa model algoritma yang optimal dalam melakukan modelling predict/klasifikasi adalah algoritma K-Nearest Neighbors dengan rentang nilai micro-average ROC-AUC dari setiap cluster sebesar 0.97 - 1.00.
